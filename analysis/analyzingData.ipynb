```python
from collections import Counter
import operator
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f=open('details_v3.csv',"r")
next(f)
data=[[],[],[]]
for l in f:
    row=l.strip().split('|#|')
  
    data[0].append(row[0])
    data[1].append(row[1])
    data[2].append(row[3])
```

```python
journals=sorted(dict(Counter(data[2])).items(),key=operator.itemgetter(1),reverse=True)
authors=sorted(dict(Counter(data[1])).items(),key=operator.itemgetter(1),reverse=True)
```

```python
sns.set()
plt.close()
plt.subplots(figsize=(10,6))
plt.plot([i for i in range(len(journals))],[i[1] for i in journals],color='b')
#plt.scatter([i for i in range(len(journals))],[i[1] for i in journals])
plt.xticks([i for i in range(len(journals))][0:0],[i[0] for i in journals][0:0],rotation=30)
plt.ylabel('Number of Articles')
plt.xlabel('Molecular Docking Journals')
plt.title('Distribution of articles in Molecular Docking Journals')
plt.show()
```

```python
plt.close()
plt.subplots(figsize=(10,6))
plt.bar([i for i in range(len(journals))][0:10],[i[1] for i in journals][0:10],color='b')
plt.xticks([i for i in range(len(journals))][0:10],[i[0] for i in journals][0:10],rotation=30)
plt.ylabel('Number of Articles')
plt.xlabel('Molecular Docking Journals')
plt.title('Top 10 Molecular Docking Journals')
plt.show()
```

```python
journals=authors
plt.close()
plt.subplots(figsize=(10,6))
plt.plot([i for i in range(len(journals))],[i[1] for i in journals],color='b')
#plt.scatter([i for i in range(len(journals))],[i[1] for i in journals])
plt.xticks([i for i in range(len(journals))][0:0],[i[0] for i in journals][0:0],rotation=30)
plt.ylabel('Number of Articles')
plt.xlabel('Molecular Docking Authors')
plt.title('Distribution of articles among Molecular Docking Authors')
plt.show()
```
